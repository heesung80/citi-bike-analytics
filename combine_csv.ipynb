```python
import glob
import pandas as pd
```

```python
path = 'city_bike' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
frame.to_csv ('conbined_citi_bike.csv', index = False, header=True)
```
